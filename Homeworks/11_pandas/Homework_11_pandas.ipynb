## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
ratings = pd.read_csv(
    'ratings.dat', 
    sep="::", 
    header=None, 
    names=['UserID', 'MovieID', 'Rating', 'Timestamp'], 
    engine='python')
ratings.tail()
```

```python
movies = pd.read_csv(
    'movies.dat', 
    sep='::', 
    header=None, 
    names=['MovieID', 'Title', 'Genres'], 
    engine='python'
)
movies.tail()
```

```python
users = pd.read_csv(
    'users.dat', 
    sep='::', 
    header=None, 
    names=['UserID', 'Gender', 'Age', 'Occupation', 'Zip-code'], 
    engine='python'
)
users.tail()
```

```python
# определяем id фильма с наибольшим количеством пятерок
movie_id_5 = (ratings[ratings.Rating == 5].
              groupby('MovieID').
              agg({'Rating' : 'count'}).
              sort_values('Rating', ascending=False).
              reset_index().
              iloc[0, 0])

# Определяем название фильма по id
movie_5 = movies[movies.MovieID == movie_id_5].iloc[0, 1]

print(f'Фильм {movie_5} имеет наибольшее количество оценок "5"')
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power_df = pd.read_csv('power.csv')
power_df.tail()
```

```python
# Смотрим названия стран
power_df.country.unique()
```

```python
baltics = ['Latvia', 'Lithuania', 'Estonia']
categories = [4, 12, 21]
query = '(country in @baltics) & (category in @categories) & (2005 <= year <= 2010) & (quantity >= 0)'
total_quantity = power_df.query(query).quantity.sum()
total_quantity
```
